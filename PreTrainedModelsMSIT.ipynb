```python
import tensorflow as tf
from tensorflow.keras.applications import InceptionV3, VGG19, ResNet101
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Model
```

```python
# InceptionV3() - 299x299x3
preModel = InceptionV3()
```

```python
preModel.summary()
```

```python
preModel = InceptionV3(include_top=False)
```

```python
# preModel.summary()
```

```python
x = preModel.output
# Adding a new hidden layer
x = Dense(1024, activation='relu')(x)
output = Dense(10, activation="softmax")(x)
```

```python
# preModel.summary()
```

```python
model = Model(inputs=preModel.input, outputs=output)
```
